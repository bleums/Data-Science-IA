#Desafío

1. Calcular el promedio de área de todos los inmuebles en los barrios en el dataset. Mostrar el top 10 y graficarlo.
2. Consultar otros datos estadísticos: conteo, mediana, valores mínimos y máximos.

```python
barrios = inmuebles.groupby([inmuebles.Barrio])
```

```python
type(barrios)
```

```python
barrios_area = barrios['Area'].mean()
barrios_area
```

```python
barrios_area.head(10).plot.bar()
```

```python
barrios_area.sort_values(ascending=False).head(10).plot.bar()
```

Los datos estadísticos más relevantes por barrio para el área de cada inmueble, como conteo, mediana, valores mínimo y máximo, se pueden obtener con la función describe()
